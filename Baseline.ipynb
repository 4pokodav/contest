# Обучение модели

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, classification_report
from sklearn.tree import DecisionTreeClassifier
from catboost import CatBoostClassifier, Pool
```

```python
train_df = pd.read_csv('train.csv')
```

```python
train_df.head()
```

```python
label_encoder = LabelEncoder()

for column in train_df.columns:
    if train_df[column].dtype == 'object':
        train_df[column] = label_encoder.fit_transform(train_df[column])
```

```python
train_df.head()
```

```python
X, y = train_df.drop(['LeaveOrNot'], axis=1).to_numpy(), train_df['LeaveOrNot'].to_numpy()
```

```python
model = CatBoostClassifier(
    iterations=100,  # Количество итераций
    learning_rate=0.1,  # Скорость обучения
    depth=8,  # Глубина дерева
    random_seed=42,
    l2_leaf_reg=5,
    verbose=0  # Отключение вывода логов
)
```

```python
model.fit(X, y)
```

# Submission

```python
test_df = pd.read_csv('test.csv')
```

```python
for column in test_df.columns:
    if test_df[column].dtype == 'object':
        test_df[column] = label_encoder.fit_transform(test_df[column])
```

```python
pred = model.predict(test_df.to_numpy())
```

```python
d = pd.DataFrame({'ID': list(range(len(pred))), 'LeaveOrNot': pred})
d.to_csv('submission.csv', index=False)
```
